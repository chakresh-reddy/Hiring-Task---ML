```python
# importing requried Libraries

import numpy as np
import pandas as pd
```

```python
# Loading Dataset

purchase = pd.read_csv("purchase.csv")
transaction = pd.read_csv("transction.csv") 
```

```python
purchase 
```

```python
transaction
```

```python
# Merge transaction data with customer purchase behavior to make single data Frame

merged = transaction.merge(purchase, on="LYLTY_CARD_NBR")

```

```python
merged
```

```python
# Identifying top 3 most profitable products

top_products = (transaction.groupby(["PROD_NBR", "PROD_NAME"])["TOT_SALES"].sum().sort_values(ascending=False))
```

```python
top_products.head(3)
```

```python
# Identifying the most loyal customer segment

loyal_customers = (merged.groupby(["LIFESTAGE", "PREMIUM_CUSTOMER"])["TOT_SALES"] .sum().sort_values(ascending=False))
```

```python
loyal_customers
```

```python
# Displaying results

print("Top 3 Most Profitable Products:\n", top_products)
print("\n Most Loyal Customers:\n", loyal_customers)
```
